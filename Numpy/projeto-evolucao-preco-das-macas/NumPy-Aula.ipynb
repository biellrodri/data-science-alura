# Regreção Linear

Onde:

a = coeficiente angular

n = número de elementos

Y = Moscow

X = datas

### Coeficiente Angular (a):

â = (n * Soma(Xi * Yi) - Soma(Xi) * Soma(Yi)) / (n * Soma(Xi^2) - (Soma(Xi))^2)

O coeficiente angular é igual a: "n" multiplicado pela soma da multiplicação entre "x" e "y", menos a soma de todos os valores de "x" multiplicado pela soma dos valores de "y". Este cálculo é dividido por "n" multiplicado pela soma do quadrado de todos os valores de "x" menos o quadrado da soma de todos os valores de "x".

### Coeficiente Linear (b):

b = Media(Yi) - â * Media(Xi)

O coeficiente linear é calculado pela média de "y" menos o coeficiente angular "a" multiplicado pela média de "x".

Lembre-se que, após calcular "a" e "b", você pode encontrar os valores de "y" usando a equação da reta:

y = a * x + b

### Calculando o coeficiente angular (a)

```python
#primeiro a gente declara as variáveis Y e X e atribui seus valores
Y = Moscow
X = datas
#aqui para ver o número de elementos dentro do array utiliza a função size()
n = np.size(Moscow)
```

```python
#para elevar elemento a elemento ao quadrado ao invés de usar a função power() que ficaria extenso demais
#é usado o **2
X**2
```

```python
#ainda para verificar se temos a mesma quantidade de elementos que o array x
#podemos envolver em ().shape e verificar
(X**2).shape
```

```python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y)) / (n*np.sum(X**2) - np.sum(X)**2)
a
```

### Calculando o coeficiente linear (b)

```python
#agora vamos atribuir o vaor de b com a expressão
b = np.mean(Y) - a*np.mean(X)
```

```python
#usando a equação da retapara achar o valor de y 
y = a*X+b
```

```python
#para conferir o valor do nosso ajuste 
np.linalg.norm(Moscow-y)
```

```python
#trazendo novamente o gráfico
plt.plot(datas, Moscow)
plt.plot(x,y)
#trazendo alguns valores para o eixo x e y
plt.plot(41.5, 41.5*a+b, '*r')
```

### Estimativa de Valor

```python
#trazendo novamente o gráfico
plt.plot(datas, Moscow)
plt.plot(x,y)
#obtendo os valores de a e b podemos estimar valores futuros de acordo com a taxa de crescimento obtido pela reta
plt.plot(41.5, 41.5*a+b, '*r')
plt.plot(100, 100*a+b, '*r')
```
