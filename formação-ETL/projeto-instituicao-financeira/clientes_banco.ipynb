## Time de Dados de uma Instituição Financeira

Através de um **arquivo CSV** que contém dados de todos os clientes cadastrados. Iremos pegar esse arquivo,

colocar em um banco de dados local e a partir dele realizar **consultas SQL**.

Assim será possível:

- **analisar dados**;

- obter **insights**;

- tomar **decisões estratégicas**

```python
#importar a biblioteca SQLALCHEMY
import sqlalchemy
```

```python
#importando algumas funções da biblioteca sqlalchemy 
from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
#criar o banco de dados
engine = create_engine('sqlite:///:memory:')
```

### **Escrevendo em um banco de dados**

```python
#variável para armazenar o endereço do arquivo csv
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/clientes_banco.csv'
```

```python
import pandas as pd
```

```python
#lendo o arquivo csv
dados = pd.read_csv(url)
```

```python
#exibindo as primeiras linhas do DataFrame
dados.head()
```

```python
#colocar as informações do DataFrame no banco de dados
dados.to_sql('clientes', engine, index=False)
```

```python
#inspencionar se o banco de dados foi criado
inspector = inspect(engine)
```

```python
#visualizar os nomes das tabelas do banco de dados
print(inspector.get_table_names())
```

```python
#criando variável para fazer uma consulta sql
query = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"'
```

```python
#ler a consulta sql
empregados = pd.read_sql(query, engine)
empregados
```

```python
#salvar essa consulta como uma nova tabela
empregados.to_sql('empregados', con=engine, index=False)
```

```python
#lendo a tabela inteira
pd.read_sql_table('empregados', engine)
```

```python
#adicionando o parâmetro de colunas
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

### **Atualizar o banco de dados**

```python
#consultando a tabela sql
query = 'SELECT * FROM clientes'
```

```python
#lendo a consulta realizada
pd.read_sql(query, engine)
```

O cliente de **ID 5008804** cancelou sua conta em nossa instituição financeira e **não faz mais parte dos nossos clientes**,
devemos **removê-lo** do nosso banco de dados

```python
#fazer uma consulta para retirar essa informação
query = text('DELETE FROM clientes WHERE ID_Cliente=5008804')
with engine.connect() as conn:
    conn.execute(query)
    conn.commit()
```

```python
#lendo a tabela atualizada
pd.read_sql_table('clientes', engine)
```

Agora precisamos atualizar a informação de **Grau_escolaridade** do cliente **ID_Cliente: 5008808**, pois ele concluiu o **ensino superior** e essa
informação deve ser **atualizada** em nosso banco de dados

```python
#realizando nova consulta
query = text('UPDATE clientes SET Grau_escolaridade="Ensino Superior" WHERE ID_Cliente=5008808')
with engine.connect() as conn:
    conn.execute(query)
    conn.commit()
```

```python
#lendo a tabela atualizada
pd.read_sql_table('clientes', engine)
```
