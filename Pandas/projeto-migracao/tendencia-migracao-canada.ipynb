## **Tendência de Migração**

Uma empresa que presta serviços de consultoria e aconselhamentos para pessoas que querem migrar do Brasil para o Canadá, quer melhorar seus serviços. Para isso, recebemos alguns arquivos disponibilizados para fazer uma análise das **tendências de migração do Brasil nos últimos anos**. E entre esses arquivos há um **arquivo csv** que contém informações de todos os países com o número aproximado de imigrantes para cada um dos anos desde 1980. Será com esse arquivo que iremos trabalhar.

```python
#importando a biblioteca pandas
import pandas as pd
```

```python
#importando o arquivo csv
df = pd.read_csv('imigrantes_canada.csv')
#visualizando o DataFrame
df
```

```python
#obter informações mais detalhadas
df.info()
```

Ao analisar as informações acima, vemos que **não há valores nulos** por tanto, não será necessário fazer um tratamento de nenhum dado nulo ou mudar algum tipo de dado.

### **Analisar as Tendências de Imigração do Brasil em um Determinado Período**

Primeiro passo a ser feito é deixar o **índex** do DataFrame como **país** para, acessar apenas o **Brasil** e extrair as informações dele

```python
#mudar o index
df.set_index('País', inplace=True)
```

```python
#criar uma variável para armazenar o intervalo  de tempo desde 1980 até 2013
anos = list(map(str, range(1980, 2014)))
```

```python
#pegando apenas os dados do Brasil
brasil = df.loc['Brasil', anos]
```

```python
#verificando os dados obtidos
brasil
```

```python
#transformando a serie acima em um DataFrame
#criando um dicionário para atribuir nome às colunas e seus valores
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes' : brasil.values.tolist()}
#utilizando o dicionário criado para criar o DataFrame
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
#visualizando o DataFrame criado
dados_brasil
```

### **Gerar o gráfico**

```python
#importando a biblioteca matplotlib
import matplotlib.pyplot as plt
```

```python
#gerando um gráfico de linha com os anos do DataFrame
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
```

### **Melhorando a Visualização do Gráfico**

```python
#alterando o tamanho do gráfico
plt.figure(figsize=(8, 4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
#adiciondo o título
plt.title('Imigração do Brasil para o Canadá')
#adicionando a descrição do eixo x
plt.xlabel('Ano')
#adicionando a descrição do eixo y
plt.ylabel('Número de imigrantes')
#aplicando um intervalo entre os anos de 5 em 5 anos
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
#exibindo o gráfico
plt.show()
```

### **Figura**

```python
#criando uma figura
fig, ax = plt.subplots(figsize=(8, 4))
#passando as informações dos eixos x e y
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
#adicionando o título à figura
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
#adicionando a descrição do eixo x
ax.set_xlabel('Ano')
#adicionando a descrição do eixo y
ax.set_ylabel('Número de imigrantes')
#adicionando um intervalo de 5 em 5 anos para o eixo x
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
#exibindo a figura
plt.show()
```

### **Boxplot**

```python
#criando a figura com dois subplots
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
#primeiro gráfico
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
#adicionando o título
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
#adicionando a descrição do eixo x
axs[0].set_xlabel('Ano')
#adicionando a descrição do eixo x
axs[0].set_ylabel('Número de imigrantes')
#inserindo um intervalo de tempo de 5 em 5 anos no eixo x
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
#adicionando um grid para visualizar melhor à qual ano corresponde cada número de imigrantes
axs[0].grid()

#segundo gráfico
axs[1].boxplot(dados_brasil['imigrantes'])
#adicionando o título
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá\n1980 a 2013')
#adicionando a descrição do eixo x
axs[1].set_xlabel('Brasil')
#adicionando a descrição do eixo x
axs[1].set_ylabel('Número de imigrantes')
#adicionando um grid para visualizar melhor à qual ano corresponde cada número de imigrantes
axs[1].grid()
plt.show()
```

### **Tendência de imigração dos 4 maiores países da América do Sul**

Agora o objetivo será criar uma figura com **quatro subplots** para analisar a **tendência de imigração** para os quatro maiores países da **América do Sul**

- **Brasil**
- **Colômbia**
- **Peru**
- **Argentina**

```python
#criando a figura
fig, axs = plt.subplots(2, 2, figsize=(10,6))
#ajustando o espaçamento entre os subplots
fig.subplots_adjust(hspace=0.5, wspace=0.3)
#adicionando um título geral
fig.suptitle('Imigração dos quatro maiores países da A')


#primeiro subplot
axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

#segundo subplot
axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

#terceiro subplot
axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

#quarto subplot
axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

#utilizando a estrutura for para adicionar intervalo de tempo de 5 em 5 anos aos eixos x dos subplots
for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))

#utilizando a estrutura for para adicionar as descrições dos eixos x e y
for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de imigrantes')

#determinando valores mínimo e máximo para o eixo y
ymin = 0
ymax = 7000

#instrução for para ajustar esses valores
for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)
    
#exibindo a figura
plt.show()
```

### **Personalizando Visualizações com Matplotlib**

Agora iremos customizar ainda mais o nosso gráfico, modificando o **tamanho de fontes**, mudando **as cores**, adicionando **mais informações**.

Para isso, começaremos mudando o **gráfico onde tínhamos apenas os dados do Brasil**. Faremos algumas modificações para **melhorar a visualização**.

```python
#criando uma figura
fig, ax = plt.subplots(figsize=(8, 4))
#passando as informações dos eixos x e y
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3,)
#adicionando o título à figura
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013', fontsize=18, loc='left')
#adicionando a descrição do eixo x
ax.set_xlabel('Ano', fontsize=14)
#adicionando a descrição do eixo y
ax.set_ylabel('Número de imigrantes', fontsize=14)
#modificando o tamanho dos ticks do eixo x
ax.xaxis.set_tick_params(labelsize=12)
#modificando o tamanho dos ticks do eixo y
ax.yaxis.set_tick_params(labelsize=12)
#adicionando um intervalo de 5 em 5 anos para o eixo x
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
#adicionando o grid
plt.grid(linestyle='--')
#exibindo a figura
plt.show()
```

### **Alterando Cores**

As cores são uma **ferramenta poderosa em visualizações de dados**. Elas podem nos ajudar a **transmitir as informações**, **destacar alguns padrões** e **identificar algumas tendências**. Algumas opções de cores incluem:

- *Base colors*;
- *Tableau palette*;
- *CSS colors*.

As **base colors** são as cores mais básicas pré-definidas e na frente do retângulo de cada cor temos a letra que representa a cor. São elas **azul(b)**, **verde(g)**, **vermelho(r)**, **ciano(c)**, **magenta(m)**, **amarelo(y)**, **preto(k)** e **branco(w)**.

A **tableau pallet** contém outras cores pré-definidas e bem distintas entre si, sendo ideal para trabalhar com **dados categóricos**.

Por fim, temos as **CSS colors**, usadas para **aplicações web** e **criação de interfaces**.

```python
#criando uma figura
fig, ax = plt.subplots(figsize=(8, 4))
#passando as informações dos eixos x e y
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3, color='g')
#adicionando o título à figura
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013', fontsize=18, loc='left')
#adicionando a descrição do eixo x
ax.set_xlabel('Ano', fontsize=14)
#adicionando a descrição do eixo y
ax.set_ylabel('Número de imigrantes', fontsize=14)
#modificando o tamanho dos ticks do eixo x
ax.xaxis.set_tick_params(labelsize=12)
#modificando o tamanho dos ticks do eixo y
ax.yaxis.set_tick_params(labelsize=12)
#adicionando um intervalo de 5 em 5 anos para o eixo x
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
#exibindo a figura
plt.show()
```

### **Comparando os dados do Brasil com todos os países da América do Sul**

Vamos avaliar o **número total de imigrantes** de cada um desses países e verificar como o **Brasil** está posicionado em relação a isso nesse período.

```python
#verificar algumas informações do nosso DataFrame
df.head()
```

Podemos ver no DataFrame que as informações são separadas por **País**, **Continente** e **Região**. Para selecionar apenas os países da América do Sul
é possível fazer uma **query** pegando somente a região que seja **América do Sul**

```python
#criando uma variável coom a query 
america_sul = df.query('Região == "América do Sul"')
```

```python
#visualizando os dados da variável criada
america_sul
```

Ao verificar o DataFrame podemos observar que há 12 países da América do Sul, e com base nisso, iremos plotar um gráfico que contenha o **número total
de imigrações** de cada um dos países. Quando falamos de gráficos que contém **quantidade**, o ideal é usar um **gráfico de barras**.

```python
#ordenando os valores do DataFrame em ordem crescente
america_sul_sorted = america_sul.sort_values('Total', ascending=True)
```

```python
#atribuindo as cores que serão utilizadas no gráfico a uma variável
cores = ['royalblue', 'orange', 'forestgreen', 'orchid',
'purple', 'brown', 'slateblue', 'gray', 'olive', 'teal',
'tomato']

#criando uma figura
fig, ax = plt.subplots(figsize=(12, 5))
#definindo os eixos da barra
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
#inserindo um título ao gráfico
ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', loc='left', fontsize=18)
#inserindo título dos rótulos
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
#tamanho dos ticks do eixo x
ax.xaxis.set_tick_params(labelsize=12)
#tamanho dos ticks do eixo y
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

No gráfico gerado acima, observamos que o Brasil está em **4º lugar**. Como nosso projeto o foco principal é em relação aos dados do **Brasil**, podemos **destacá-lo** das demais colunas

```python
#criando uma lista de cores vazia
cores = []
#criando condição for
for pais in america_sul_sorted.index:
    if pais == 'Brasil':
        cores.append('green')
    else:
        cores.append('silver')


#criando uma figura
fig, ax = plt.subplots(figsize=(12, 5))
#definindo os eixos da barra
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
#inserindo um título ao gráfico
ax.set_title('América do Sul: Brasil foi o 4º país com mais imigrantes\npara o Canadá no período de 1980 a 2013', loc='left', fontsize=18)
#inserindo título dos rótulos
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
#tamanho dos ticks do eixo x
ax.xaxis.set_tick_params(labelsize=12)
#tamanho dos ticks do eixo y
ax.yaxis.set_tick_params(labelsize=12)

#criando uma condição para acrescentar texto às barras
for i, v in enumerate(america_sul_sorted['Total']):
    ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

#retirando o frame do gráfico
ax.set_frame_on(False)

#retirando a vissualização do eixo x
ax.get_xaxis().set_visible(False)

#retirando os marcadores dos eixos
ax.tick_params(axis='both', which='both', length=0)

plt.show()
```
