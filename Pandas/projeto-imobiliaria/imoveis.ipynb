## **Manipulando os Dados**

O time de desenvolvimento do site da empresa solicitou criação de duas novas colunas numéricas na nossa base de dados.
No site dos imóveis, eles gostariam de apresentar informações mais sumarizadas em relação a valores. Para isso, essas
colunas serão importantes:

- **valor por mês** essa coluna deve conter os gastos mensais de cada imóvel,
incluindo aluguel e condomínio;

- **valor por ano** essa coluna deve conter os gastos anuais por imóvel, ou seja,
IPTU mais 12 meses de aluguel e condomínio.

```python
url = 'aluguel.csv'
dados = pd.read_csv('aluguel.csv', sep=';')
```

```python
dados.head()
```

### Criando as colunas numéricas

```python
#criando coluna de valor por mês
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

```python
#criando coluna de valor por ano
dados['Valor_por_ano'] = dados['Valor'] * 12 + dados['IPTU']
dados.head()
```

## **Colunas Categóricas**

O time de desenvolvimento solicitou a criação de mais duas colunas. No entanto,
dessa vez são categóricas:

- **Descrição**: essa coluna deve possuir uma sumarização das principais informações dos imóveis
que serão apresentadas no site: *tipo de imóvel*, *bairro*, *quantidade de quartos* e *vagas de garagem*;

- **Possui_suite**: essa deve ser uma coluna que informe apenas se o imóvel possui ou não suítes,
sem se importar com a quantidade.

### Criando a coluna descrição

```python
#concatenar as linhas da coluna Tipo com as linhas da coluna Bairro
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro']
dados.head()
```

```python
#adicionando a quantidade de quartos e a quantidade de vagas
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                     dados['Quartos'].astype(str) + ' quarto(s) ' + \
                     ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem. '
dados.head() 
```

### Criando a coluna Possui_suite

```python
#adicionando a coluna Possui_suite
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dados.head()
```

```python
#salvando os dados em um arquivo csv para o time de desenvolvimento
dados.to_csv('dados_completos_dev.csv', index=False, sep=';')
```
